```python
import numpy as np

```

```python
W = np.matrix([[0.6, 0.7 , 0], [0.01, 0.43, 0.88]])
w = np.matrix([[0.02], [0.03],[0.09]])
X = np.matrix([[0.75, 0.8],[0.2,0.05],[-0.75, 0.8],[0.2, -0.05]])
Y = np.matrix([[1],[1],[-1],[-1]])

for i in range(0,3):
    s = np.dot(X,W)
    print "s"
    print s

    sReLU = s.clip(min=0)
    print "\n sRelu"
    print sReLU

    sout = np.dot(sReLU,w)
    print "\n Sout"
    print sout

    yout = np.tanh(sout)
    print "\n yout"
    print yout

    Loss = 0.5*np.power((yout - Y),2)
    print "\n Loss"
    print Loss

    dSout = np.multiply(1 - np.power(np.tanh(sout),2),(yout - Y))
    print "\n dSout"
    print dSout

    #derivative of ReLU
    sReluDeriv = sReLU
    sReluDeriv[sReluDeriv>0] = 1
    D = np.multiply(sReluDeriv, np.dot(dSout, w.T))
    print "\n D"
    print D

    dW = np.dot(X.T, D)
    print "\n dW"
    print dW
    
    dWout = np.dot(sReLU.T, dSout)
    print "\n dWout"
    print dWout
    
    print "\n new W"
    W = W - 0.5*dW
    print W

    print "\n new Wout"
    w = w - 0.5*dWout
    print w
    print "=================="
```
